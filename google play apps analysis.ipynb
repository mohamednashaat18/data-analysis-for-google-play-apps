# pre_processing data anaysis

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
#I will transform size(KB) to size(MB)
df['size(kb)']=df['size(kb)']/1000
df=df.rename(columns={'size(kb)':'size(MB)'})
df.head()
```
